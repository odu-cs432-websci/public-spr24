# CS 432/532 - NetworkX Example

NetworkX: http://networkx.github.com/ 

Has Karate Club network as a built-in function (`karate_club_graph()`)

The example below is based on https://networkx.github.io/documentation/stable/auto_examples/graph/plot_karate_club.html

```
import matplotlib.pyplot as plt
import networkx as nx
```

Generate the Zachary's Karate Club graph as G

```
G = nx.karate_club_graph()
```

Print each node and its degree

```
print("Node\tDegree")
for v in G:
    print('{}\t{}'.format(v, G.degree(v)))
```

Color nodes 0 (Mr. Hi) and 33 (John). These are the most connected nodes (highest degrees).

*Default color (#1f78b4) for nodes found at https://networkx.github.io/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html*

```
color = ["#1f78b4"] * 34
color[0] = "green"  # Mr. Hi
color[33] = "red"   # John
```

Draw the graph with a circular layout with node labels and Mr. Hi and John's nodes colored.

```
nx.draw_circular(G, with_labels=True, node_color=color)
# plt.show()   # needed if not in the notebook
```

Draw the graph with a spring layout

```
nx.draw_spring(G, with_labels=True, node_color=color)
# plt.show()   # needed if not in the notebook
```

Draw with force-directed layout

```
nx.draw_kamada_kawai(G, with_labels=True, node_color=color)
# plt.show()   # needed if not in the notebook
```

Delete the only edge connecting node 11 to the rest

```
G.remove_edge(11, 0)
```

Draw with the force-directed layout (kamada_kawai).

```
nx.draw_kamada_kawai(G, with_labels=True, node_color=color)
```

Draw the same graph with the spring layout.  Notice how the disconnected Node 11 is separate from the other nodes (*this will be important for HW5*).

```

nx.draw_spring(G, with_labels=True, node_color=color)
# plt.show()   # needed if not in the notebook
```
